```python
# График улыбок волатильности в модели CEV (лекции 7)

%load_ext autoreload
%autoreload 2
from math import *
import numpy as np
import scipy.stats as st
import matplotlib as mpl
import matplotlib.pyplot as plt

import sys
sys.path.append('../..')

import warnings
warnings.filterwarnings('ignore', category=FutureWarning)

import pyfinmath as fm

plt.rcParams.update({
    "font.size": 11,
    "axes.titlesize": 8,
    "axes.labelsize": 8,
    "xtick.labelsize": 7,
    "ytick.labelsize": 7,
    "legend.fontsize": 7,
    "figure.titlesize": 8,
    "figure.dpi": 600,
    "lines.linewidth": 1,
    "axes.linewidth": 0.6,
    "grid.linewidth": 0.3,
    "xtick.major.width": 0.6,
    "ytick.major.width": 0.6,
    "xtick.minor.width": 0.3,
    "ytick.minor.width": 0.3,
})

scale = 1     # Масштаб графиков (меньше 1, чтобы помещались на А4)
```

```python
# Улыбки волатильности в модели CEV
fig, ax = plt.subplots(1, 1, figsize=(2.7*scale, 1.8*scale), constrained_layout=True, sharey=True)
ax.set_xlabel('Страйк')
ax.set_ylabel('Волатильность')

K = np.linspace(0.50, 1.8, 100)
for beta in [0.25, 0.5, 0.75, 0.9]:
    model = fm.CEV(0.3, beta)
    ax.plot(K, model.implied_vol(1, maturity=1, strike=K), label=f"$\\beta={beta}$")

ax.set_ylim(0.19,0.41)
ax.legend(frameon=False)
```
